```python
def demo():
    test_file_path='/raidHDD/experimentData/money/test/TESTALL'
    test_file=listdir(test_file_path)
    for file in test_file:
        image_location=os.path.join(test_file_path,file)
        image=cv2.imread(image_location)
        image_for_show=copy.deepcopy(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        plt.imshow(image_for_show)
        plt.show()
        image=image-image.mean()
        image[image<0]=0
        image=cv2.resize(image,(64,64))
        image=image.astype(np.uint8)
        image=image/128
    
        result=model.predict_yours([image])
        label=['10c','1c','25c']
        
        print('This is '+label[result[0]]+'!')
```
